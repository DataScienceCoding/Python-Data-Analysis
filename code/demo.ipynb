首先，使用以下命令导入数据集sklearn

```python
from sklearn.datasets import load_iris
from sklearn.datasets import load_boston
```

然后，导入Matplotlib绘图模块

```python
import matplotlib
from matplotlib import pyplot as plt
# 内嵌画图
%matplotlib inline
```

加载iris数据集，显示数据集的相关描述，同时将第1列（萼片长度）作为x坐标值，将第2列（萼片宽度）作为y坐标值

```python
irirs = load_iris()
print(irirs.DESCR)
data = irirs.data
plt.plot(data[:, 0], data[:, 1], ".")
```

加载boston数据集，显示数据集的相关描述，同时将第3列（非零售业务的比例）作为x坐标值，将第5列（一氧化氮浓度）作为y坐标值，图上每个点用“+”号表示。

```python
boston = load_boston()
print(boston.DESCR)
data = boston.data
plt.plot(data[:, 2], data[:, 4], "+")
```
